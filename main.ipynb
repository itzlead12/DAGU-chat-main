```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator


def qubit1(bits, bases):
    circuit1 = []
    for bit, basis in zip(bits, bases):
        qc = QuantumCircuit(1, 1)
        if bit==1:
            qc.x(0)
        if basis=='X':
            qc.h(0)
        circuit1.append(qc)
    return circuit1

def qubit2(bits, bases):
    circuit2 =[]
    for qc, basis in zip(bits, bases):
        qc1 = qc.copy()
        if basis=='X':
            qc1.h(0)
        qc1.measure(0, 0)
        circuit2.append(qc1)
    return circuit2

def qubit3(bits, bases):
    circuit3 =[]
    for qc1, basis in zip(bits, bases):
        qc2 = qc1.copy()
        if basis=='X':
            qc2.h(0)
        qc2.measure(0, 0)
        circuit3.append(qc2)
    return circuit3

def abebebitbase(qubitno):
    bits =[]
    basis =[]
    for i in range(qubitno):
        while True:
            cls=input(f"Abebe, Enter the value of qubit {i} (0 or 1): ").strip()
            if cls in ['0', '1']:
                bits.append(int(cls))
                break
            else:
                print("Invalid input")
        while True:
            clb=input(f"Abebe, Enter the basis for qubit {i} (Z or X): ").strip().upper()
            if clb in ['Z', 'X']:
                basis.append(clb)
                break
            else:
                print("Invalid input")
    return bits, basis

def hackerbase(qubitno):
    basis3 =[]
    for i in range(qubitno):
        while True:
            clh=input(f"Hacker, Enter the basis measurement for qubit {i} (Z or X): ").strip().upper()
            if clh in ['Z', 'X']:
                basis3.append(clh)
                break
            else:
                print("Invalid input")
    return basis3

def kebedebase(qubitno):
    basis2 =[]
    for i in range(qubitno):
        while True:
            clk=input(f"Kebede, Enter the basis measurement for qubit {i} (Z or X): ").strip().upper()
            if clk in ['X', 'Z']:
                basis2.append(clk)
                break
            else:
                print("Invalid input")
    return basis2

def main():
    qubitno =int(input("Enter the amount of bit to send: "))
    abebebit, abebebasis = abebebitbase(qubitno)
    hackerbasis= hackerbase(qubitno)
    kebedebasis = kebedebase(qubitno)

    abebe = qubit1(abebebit, abebebasis)
    hacker = qubit2(abebe, hackerbasis)
    kebede = qubit3(hacker, kebedebasis)

    simulator = AerSimulator()
    kebedebit = []
    print("\nResults: ")
    for i, qc in enumerate(kebede):
        result = simulator.run(qc, shots=100).result()
        counts = result.get_counts()
        if len(counts) == 1 and list(counts.values())[0] == 100:
            measured = int(list(counts.keys())[0])
            kebedebit.append(measured)
        else:
            kebedebit.append(None)

        print(f"QUBIT {i}:   Abebe bit:{abebebit[i]} Kebede bit:{counts} Abebe basis:{abebebasis[i]} Kebede basis:{kebedebasis[i]} Hacker basis: {hackerbasis[i]}")

    basematch = [i for i in range(qubitno) if abebebasis[i] == kebedebasis[i]]
    errors = sum(1 for i in basematch if abebebit[i] != kebedebit[i])
    errorrate = errors / len(basematch) if basematch else 0
    if len([abebebit[i] for i in basematch])==0:
        print("No key generated. Try again")
    else:
        print("The main key is:", [abebebit[i] for i in basematch])
        print("The checking key is:", [abebebit[i] for i in basematch if abebebit[i] == kebedebit[i]])
        print(f"The error rate is: {errorrate:.2f}")

        if errorrate == 0.0:
            print("the key is safe, You can use it.")
        else:
            print("the key is not safe, There might be hacker.")

if __name__=="__main__":
    main()
```
